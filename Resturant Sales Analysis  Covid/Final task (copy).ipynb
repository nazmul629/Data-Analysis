###  People response retio  

```python
number_of_people= df.groupby('Period')['Avg. no of customers per day'].sum()
number_of_people
```

```python
percentagesOfPeople =  number_of_people / df["Avg. no of customers per day"].sum()*100
percentagesOfPeople
```

```python
labels = ['Before covid19', 'After covid19']
percentages = percentagesOfPeople
plt.pie(percentages, labels=labels, autopct='%1.1f%%')
plt.title('Percentage Comparison Before and After Covid19')
plt.show()
```

### Chicken Item  Analisis 

```python
chicken_grill_sale = df.groupby('Period')['chicken Grill'].sum()
chicken_grill_sale
```

```python
chicken_grillPercentages = chicken_grill_sale/number_of_people * 100
chicken_grillPercentages
```

```python

categories = ['Before covid19', 'After covid19']
percentages = chicken_grillPercentages

# Create the bar plot
plt.bar(categories, percentages)
plt.xlabel('Time')
plt.ylabel('Percentage')
plt.title('Percentage Comparison Before and After Covid19')
plt.show()
```

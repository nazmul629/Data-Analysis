# Data analysis and visualizations

```python
import matplotlib.pyplot as plt

# Cancelration ratio 
cancelled_perc = df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)
plt.figure(figsize = (5,4))

plt.title("Resrvation Status Count")
plt.bar(['Not Canceled', 'Canceled'],df["is_canceled"].value_counts(), color=['g','r'], width = .7)
plt.show()
```

```python
# Cancelration ratio  each hotel 
plt.figure(figsize=(8,4))
ax1 = sns.countplot(x ='hotel', hue ='is_canceled', data = df, palette='Blues' )

# ax1.legend(bbox_to_anchor(1,1))
plt.title("Reservation Status in different in hotels",size=20)
plt.xlabel("hotel")
plt.ylabel("Number of reservations")
plt.show()
```

```python
#Resourt hotel Ratio 

resort_hotel = df[df['hotel']=="Resort Hotel"]
resourt_hotel_caneld_parc=resort_hotel['is_canceled'].value_counts(normalize=True)

print(resourt_hotel_caneld_parc)

plt.figure(figsize = (5,4))

plt.title("Resort Hotel Resrvation Stats Count",fontsize=20)
plt.bar(['Not Canceled', 'Canceled'],resort_hotel["is_canceled"].value_counts(), color=['g','r'], width = .7)

plt.show()
```

```python
#Resourt hotel Ratio 

city_hotel = df[df['hotel']=="City Hotel"]
city_hotel_caneld_parc=city_hotel['is_canceled'].value_counts(normalize=True)

print(city_hotel_caneld_parc)

plt.figure(figsize = (5,4))

plt.title("City Hotel Resrvation Status Count",fontsize=20)
plt.bar(['Not Canceled', 'Canceled'],city_hotel["is_canceled"].value_counts(), color=['g','r'], width = .7)

plt.show()
```

```python
resort_hotel = resort_hotel.groupby("reservation_status_date")[['adr']].mean()
city_hotel = city_hotel.groupby("reservation_status_date")[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index,resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```
